```python
from models_.building.pubmed_tokenizer import load_article_tokenizer as load_tokenizer
from data.pubmed.from_json import FromJsonDataset
from data.pubmed.tokenized import TokenizedDataset

from storage.sqlite.storage_ import SqliteStorage
```

```python
dataset = FromJsonDataset(json_file="../data/pubmed_500K.json")
```

```python
tokenizer = load_tokenizer()
tokenizer
```

```python
MAX_TOKEN_LENGTH = 800 # check train_dataset.ipynb for more details
```

```python
tokenized_dataset = TokenizedDataset(
    tokenizer=tokenizer,
    dataset=dataset,
    max_length=MAX_TOKEN_LENGTH,
)
```

```python
tokenized_dataset[0]
```

```python
sqlite_storage = SqliteStorage(
    dimension=MAX_TOKEN_LENGTH,
    db_path="../outputs/store/pubmed_500K.db",
)
```

```python
for i in range(0, 1000):
    sqlite_storage.store(
        [(dataset[i]["id"], tokenized_dataset[i]["input_ids"][0].numpy().astype("float32"), dataset[i]["contents"])],
    )
```
